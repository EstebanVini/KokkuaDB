```python
import pandas as pd


df = pd.read_csv('./Datasets/Proveedores.csv', dtype=str)
```

```python
df
```

```python
from faker import Faker

fake = Faker(['es_MX'])


# Getting all unique values of establecimiento and only leave the row with its first appearance
df = df.drop_duplicates(subset=['ESTABLECIMIENTO'], keep='first')
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
# Remove the scape sequences from all string columns
def remove_scape(column):
    return column.str.replace('\n', '').str.replace('\r', '').str.replace('\t', '')


df['ESTABLECIMIENTO'] = remove_scape(df['ESTABLECIMIENTO'])
df['COLONIA'] = remove_scape(df['COLONIA'])
df['CALLE No.'] = remove_scape(df['CALLE No.'])
```

```python
df
```

```python
# Concatenate the columnns CALLE No.	COLONIA	CÓDIGO\nPOSTAL	LOCALIDAD	ENTIDAD into one column named DIRECCION

df['DIRECCION'] = df['CALLE No.'] + ', ' + df['COLONIA'] + ', ' + df['CÓDIGO\nPOSTAL'] + ', ' + df['LOCALIDAD'] + ', ' + df['ENTIDAD']

# Drop the columns that are not needed anymore
df = df.drop(columns=['CALLE No.', 'COLONIA', 'CÓDIGO\nPOSTAL', 'LOCALIDAD', 'ENTIDAD'])
```

```python
df
```

```python
# only leave establecimiento and direccion columns
df = df[['ESTABLECIMIENTO', 'DIRECCION']]
```

```python
df
```

```python
phoneNumbers = pd.read_json('./Datasets/phoneNumbers.json')
```

```python
# Change rename the first column to telefono, use index because the column name is a number
phoneNumbers.rename(columns={phoneNumbers.columns[0]: 'TELEFONO'}, inplace=True)

```

```python
phoneNumbers
```

```python
# Add a column to the dataframe with the phone numbers
df['TELEFONO'] = phoneNumbers['TELEFONO']

```

```python
df.to_csv('./Datasets/ProveedoresLimpio.csv', index=False)
```

```python
import mysql.connector

from dotenv import load_dotenv
import os
load_dotenv()

DB_USER = os.getenv('MYSQL_USER')
DB_PASSWORD = os.getenv('MYSQL_PASSWORD')
DB_HOST = os.getenv('MYSQL_HOST')
DB_DATABASE = os.getenv('MYSQL_DATABASE')




cnx = mysql.connector.connect(user=DB_USER, password=DB_PASSWORD,
                                host=DB_HOST,
                                database=DB_DATABASE)
```

```python

# Insert the data into the database table Proveedores

cursor = cnx.cursor()

for index, row in df.iterrows():
    cursor.execute("INSERT INTO Proveedores (Nombre, Ubicación, NumContacto) VALUES (%s, %s, %s)", (row['ESTABLECIMIENTO'], row['DIRECCION'], row['TELEFONO']))
    
cnx.commit()

cursor.close()

cnx.close()
```
